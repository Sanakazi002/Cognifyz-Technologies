```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 7))
res.dropna(subset=['Latitude','Longitude'],inplace=True)
ax = plt.axes(projection=ccrs.PlateCarree())
plt.scatter(res['Longitude'],res['Latitude'], color='blue', marker='^', s=100, transform=ccrs.PlateCarree())
ax.coastlines()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Prepare the data for KMeans clustering (Latitude and Longitude values)
X = res[['Latitude', 'Longitude']].values
# Calculate inertia for a range of K values (e.g., from 1 to 10)
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
# Plot the inertia for each K value
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
res.dropna(subset=['Latitude','Longitude'],inplace=True)
X = res[['Latitude', 'Longitude']].values
kmeans = KMeans(n_clusters=2, random_state=0)
res['cluster'] = kmeans.fit_predict(X)
plt.figure(figsize=(7,4))
Scatter=plt.scatter(res['Longitude'], res['Latitude'], c=res['cluster'], cmap='viridis', marker='o', s=50)
plt.legend(handles=Scatter.legend_elements()[0], labels=['Cluster 1', 'Cluster 2'], title="Clusters")
plt.title('Clusters of Restaurant in specific areas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
